```python
#@title Type the word that you want to translate in many languages

# print(LANGUAGES_TRANSLATE)

#https://colab.research.google.com/notebooks/snippets/advanced_outputs.ipynb
LIST_LANGUAGES = []

#Not in Use
for k, l in LANGUAGES_TRANSLATE.items():
  LIST_LANGUAGES.append(l)

# print(LIST_LANGUAGES)

#@markdown Text Translated in 7 Languages, include English, Portuguese, Thai, Spanish, French, Polish, Italian

#@markdown Future with selection of Languages to Translate 

#@markdown English is the default languages
text_to_translate = "olha o que eu fiz" #@param {type:"string"}

#select_or_input = "apples" #@param ["apples", "bananas", "oranges"] {allow-input: true}
#from_Language =  "english" #@param ['afrikaans', 'albanian', 'amharic', 'arabic', 'armenian', 'azerbaijani', 'basque', 'belarusian', 'bengali', 'bosnian', 'bulgarian', 'catalan', 'cebuano', 'chichewa', 'chinese (simplified)', 'chinese (traditional)', 'corsican', 'croatian', 'czech', 'danish', 'dutch', 'english', 'esperanto', 'estonian', 'filipino', 'finnish', 'french', 'frisian', 'galician', 'georgian', 'german', 'greek', 'gujarati', 'haitian creole', 'hausa', 'hawaiian', 'hebrew', 'hebrew', 'hindi', 'hmong', 'hungarian', 'icelandic', 'igbo', 'indonesian', 'irish', 'italian', 'japanese', 'javanese', 'kannada', 'kazakh', 'khmer', 'korean', 'kurdish (kurmanji)', 'kyrgyz', 'lao', 'latin', 'latvian', 'lithuanian', 'luxembourgish', 'macedonian', 'malagasy', 'malay', 'malayalam', 'maltese', 'maori', 'marathi', 'mongolian', 'myanmar (burmese)', 'nepali', 'norwegian', 'odia', 'pashto', 'persian', 'polish', 'portuguese', 'punjabi', 'romanian', 'russian', 'samoan', 'scots gaelic', 'serbian', 'sesotho', 'shona', 'sindhi', 'sinhala', 'slovak', 'slovenian', 'somali', 'spanish', 'sundanese', 'swahili', 'swedish', 'tajik', 'tamil', 'tatar', 'telugu', 'thai', 'turkish', 'turkmen', 'ukrainian', 'urdu', 'uyghur', 'uzbek', 'vietnamese', 'welsh', 'xhosa', 'yiddish', 'yoruba', 'zulu'] {allow-input: false}

#Instance of Google Translator
import requests 
from google_trans_new.google_trans_new import google_translator  
import gtts

import librosa
import IPython.display as ipd
import warnings
warnings.filterwarnings('ignore') 
translator = google_translator() 

#Testing With Player
filename = "translation_lang.mp3"

#Translation English <> English (OBS: Don't use the same Language with SRC and TGT)
language = translator.detect(text_to_translate)
print("Translate text {} from {}\n".format(text_to_translate, language[1].upper())) 
#Translation English <> Portuguese
translation = translator.translate(text_to_translate, lang_src='auto',lang_tgt='en',pronounce=True)  
print("English: {}".format(translation[0]))
lang = "en"
# make request to google to get synthesis
tts = gtts.gTTS(translation[0], lang=lang)
# Saving file and Playing
tts.save(filename.replace("_lang","_" + lang))
x, sr = librosa.load(filename.replace("_lang","_" + lang)) 
pa = ipd.Audio(x, rate=sr)
display(pa)
#Translation English <> Portuguese
translation = translator.translate(text_to_translate, lang_src='auto',lang_tgt='pt',pronounce=True)  
print("Portugues: {}".format(translation[0]))
lang = "pt"
# make request to google to get synthesis
tts = gtts.gTTS(translation[0], lang=lang)
# Saving file and Playing
tts.save(filename.replace("_lang","_" + lang))
x, sr = librosa.load(filename.replace("_lang","_" + lang)) 
pa = ipd.Audio(x, rate=sr)
display(pa)
#Translation English <> Thai
translation = translator.translate(text_to_translate, lang_src='auto',lang_tgt='th',pronounce=True)  
print("Thai: {} ({})".format(translation[0],translation[2]))
lang = "th"
# make request to google to get synthesis
tts = gtts.gTTS(translation[0], lang=lang)
# Saving file and Playing
tts.save(filename.replace("_lang","_" + lang))
x, sr = librosa.load(filename.replace("_lang","_" + lang)) 
pa = ipd.Audio(x, rate=sr)
display(pa)
#Translation English <> Spanish
translator = google_translator() 
translation = translator.translate(text_to_translate, lang_src='auto',lang_tgt='es',pronounce=True)  
print("Spanish: {}".format(translation[0]))
lang = "es"
# make request to google to get synthesis
tts = gtts.gTTS(translation[0], lang=lang)
# Saving file and Playing
tts.save(filename.replace("_lang","_" + lang))
x, sr = librosa.load(filename.replace("_lang","_" + lang)) 
pa = ipd.Audio(x, rate=sr)
display(pa)
#Translation English <> French
translation = translator.translate(text_to_translate, lang_src='auto',lang_tgt='fr',pronounce=True)  
print("French: {}".format(translation[0]))
lang = "fr"
# make request to google to get synthesis
tts = gtts.gTTS(translation[0], lang=lang)
# Saving file and Playing
tts.save(filename.replace("_lang","_" + lang))
x, sr = librosa.load(filename.replace("_lang","_" + lang)) 
pa = ipd.Audio(x, rate=sr)
display(pa)
#Translation English <> Polish
translation = translator.translate(text_to_translate, lang_src='auto',lang_tgt='pl',pronounce=True)  
print("Polish: {}".format(translation[0]))
lang = "pl"
# make request to google to get synthesis
tts = gtts.gTTS(translation[0], lang=lang)
# Saving file and Playing
tts.save(filename.replace("_lang","_" + lang))
x, sr = librosa.load(filename.replace("_lang","_" + lang)) 
pa = ipd.Audio(x, rate=sr)
display(pa)
#Translation English <> Polish
translation = translator.translate(text_to_translate, lang_src='auto',lang_tgt='it',pronounce=True)  
print("Italian: {}".format(translation[0]))
lang = "it"
# make request to google to get synthesis
tts = gtts.gTTS(translation[0], lang=lang)
# Saving file and Playing
tts.save(filename.replace("_lang","_" + lang))
x, sr = librosa.load(filename.replace("_lang","_" + lang)) 
pa = ipd.Audio(x, rate=sr)
display(pa)
```
